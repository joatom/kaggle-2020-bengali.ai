```python
import pandas as pd
import numpy as np
from scipy import stats

import sklearn.metrics
```

```python
df_labels = pd.read_csv('input/train.csv').drop('grapheme',axis=1)

df_labels
```

```python
for i in range(3):
    s = str(i+1)

    df=pd.read_csv('mod'+s+'_train_inference')
    df.columns=['image_id','gr'+s,'vd'+s,'cd'+s]
    
    df_labels = df_labels.merge(df,on='image_id')

# mode
```

```python
y = df_labels[['grapheme_root','vowel_diacritic','consonant_diacritic']].values
y_pred1 = df_labels[['gr1','vd1','cd1']].values
y_pred2 = df_labels[['gr2','vd2','cd2']].values
y_pred3 = df_labels[['gr3','vd3','cd3']].values
y
```

```python
def scores(y, y_pred):
    rs1=sklearn.metrics.recall_score(y[:,0], y_pred[:,0], average='macro')
    rs2=sklearn.metrics.recall_score(y[:,1], y_pred[:,1], average='macro')
    rs3=sklearn.metrics.recall_score(y[:,2], y_pred[:,2], average='macro')

    return rs1, rs2, rs3, np.average([rs1, rs2, rs3], weights=[2,1,1])
```

```python
print('score mod1:',scores(y, y_pred1))
print('score mod2:',scores(y, y_pred2))
print('score mod3:',scores(y, y_pred3))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics  import recall_score
import pickle
```

```python
df_gr = df_labels[['gr1','gr2','gr3']].values
y_gr = df_labels[['grapheme_root']].values.ravel() #[:-20000]
df_vd = df_labels[['vd1','vd2','vd3']].values
y_vd = df_labels[['vowel_diacritic']].values.ravel() #[:-20000]
df_cd = df_labels[['cd1','cd2','cd3']].values
y_cd = df_labels[['consonant_diacritic']].values.ravel() #[:-20000]

df_all = df_labels[['gr1','gr2','gr3','vd1','vd2','vd3','cd1','cd2','cd3']].values #[:-20000]
```

```python
df_all.shape, y_gr.shape
```

```python
modGR = RandomForestClassifier(random_state=2020).fit(X=df_all, y=y_gr)
modVD = RandomForestClassifier(random_state=2020).fit(X=df_all, y=y_vd)
modCD = RandomForestClassifier(random_state=2020).fit(X=df_all, y=y_cd)
```

```python
pickle.dump(modGR, open('RF_gr_all_100', 'wb'))
pickle.dump(modVD, open('RF_vd_all_100', 'wb'))
pickle.dump(modCD, open('RF_cd_all_100', 'wb'))
```

```python
preds_gr = modGR.predict(df_all)
preds_vd = modVD.predict(df_all)
preds_cd = modCD.predict(df_all)
```

```python
# all 90 %
```

```python
rfrs1=recall_score(y_gr, preds_gr, average='macro')
rfrs2=recall_score(y_vd, preds_vd, average='macro')
rfrs3=recall_score(y_cd, preds_cd, average='macro')

rfrs1, rfrs2, rfrs3,np.average([rfrs1, rfrs2, rfrs3], weights=[2,1,1])
```
